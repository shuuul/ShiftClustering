```python
from time import time

from sklearn.cluster import MeanShift, DBSCAN
from sklearn.datasets import make_blobs

from shiftclustering import MeanShiftPP, GridShift
```

```python

X, labels_true = make_blobs(n_samples=500, n_features=4, centers=10, random_state=42)
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
# using t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(8, 8))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels_true, cmap='viridis', s=5)
plt.title('t-SNE visualization of Ground-truth')
plt.show()
```

```python
# Initialize the clustering algorithms
meanshift = MeanShift(bandwidth=2.0, max_iter=1000)
meanshiftpp = MeanShiftPP(bandwidth=1.0, max_iter=1000)
gridshift = GridShift(bandwidth=1.0, max_iters=1000)
```

```python
labels_meanshiftpp = meanshiftpp.fit_predict(X)
labels_meanshift = meanshift.fit_predict(X)
labels_gridshift = gridshift.fit_predict(X)
```

```python
from sklearn import metrics

ARI_meanshift = metrics.adjusted_rand_score(labels_true, labels_meanshift)
NMI_meanshift = metrics.normalized_mutual_info_score(labels_true, labels_meanshift)

ARI_meanshift, NMI_meanshift
```

```python
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Visualizing the true labels
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels_true, cmap='viridis', s=5)
plt.title('t-SNE visualization (True labels)')

# Visualizing the predicted labels
plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels_meanshift, cmap='viridis', s=5)
plt.title('t-SNE visualization (Mean-Shift Predicted labels)')
```

```python
ARI_meanshiftpp = metrics.adjusted_rand_score(labels_true, labels_meanshiftpp)
NMI_meanshiftpp = metrics.normalized_mutual_info_score(labels_true, labels_meanshiftpp)

ARI_meanshiftpp, NMI_meanshiftpp
```

```python
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Visualizing the true labels
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels_true, cmap='viridis', s=5)
plt.title('t-SNE visualization (True labels)')

# Visualizing the predicted labels
plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels_meanshiftpp, cmap='viridis', s=5)
plt.title('t-SNE visualization (Mean-Shift++ Predicted labels)')
```

```python
ARI_gridshift = metrics.adjusted_rand_score(labels_true, labels_gridshift)
NMI_gridshift = metrics.normalized_mutual_info_score(labels_true, labels_gridshift)

ARI_meanshiftpp, NMI_meanshiftpp
```
